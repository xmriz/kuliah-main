## **Membaca Data**

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('H05.csv')
```

## **Mengakses Data**

```python
# Histogram distribusi nilai matematika
df['nilai_mat'].plot(kind='hist', title='Distribusi Nilai Matematika')
```

```python
# Diagram batang horizontal banyaknya mahasiswa masing-masing fakultas
df['fakultas'].value_counts().plot(kind='barh', title='Frekuensi Jumlah Mahasiswa tiap Fakultas')
```

```python
# Diagram pie banyaknya mahasiswa masing-masing fakultas
df['fakultas'].value_counts().plot(kind='pie', title='Frekuensi Jumlah Mahasiswa tiap Fakultas')
```

```python
# Berdasarkan diagram pie dan horizontal diatas, dapat dilihat fakultas dengan mahasiswa terbanyak yaitu FTI

# Diagram batang horizntal lebih baik dibandingkan dengan diagram pie, 
# karena dengan menggunakan diagram batang horizontal kita dapat mengetahui perkiraan nilai dari jumlah mahasiswa di suatu fakultas, 
# sedangkan, diagram pie hanya memperlihatkan rasio jumlah mahasiswa pada suatu fakultas terhadap total jumlah mahasiswa keseluruhan
```

```python
# Scatter plot dengan nilai kimia sebagai x dan fisika sebagai y
df.plot(kind='scatter', x='nilai_kim', y='nilai_fis', title='nilai kimia VS nilai fisika')
```
