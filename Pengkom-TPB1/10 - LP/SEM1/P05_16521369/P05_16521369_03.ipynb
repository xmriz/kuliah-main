## **Membaca Data**

```python
import pandas as pd
import matplotlib as plt

df = pd.read_csv('P05-01.csv')
```

## **Mengakses Data**

```python
# Jawaban 1
# Histogram distribusi total nominal yang belum dibayar pengguna dengan status aktif(Existing Customer), dengan bins 20

df.loc[df['Attrition_Flag'] == 'Existing Customer']['Total_Revolving_Bal'].plot(kind = 'hist', title = 'total nominal yang belum dibayar pengguna dengan status aktif', bins=20)
```

```python
# Jawaban 2
# Pie chart perbandingan banyaknya masing-masing pendidikan terakhir dari pengguna dengan banyak transaksi selama 12 bulan terakhir di atas 30

df.loc[df['Total_Trans_Ct'] > 30]['Education_Level'].value_counts().plot(kind='pie', title='banyaknya masing-masing pendidikan terakhir dari pengguna dengan banyak transaksi selama 12 bulan terakhir di atas 30')
```

```python
# Jawaban 3
# Scatter plot dengan usia pengguna sebagai x dan nominal limit kartu kredit sebgai y dari pengguna berjenis kelamis laki-laki

df.loc[df['Gender'] == 'M'].plot(kind='scatter', x='Customer_Age', y='Credit_Limit', title = 'Scatter plot dengan usia pengguna sebagai x dan nominal limit kartu kredit sebgai y dari pengguna berjenis kelamis laki-laki')
```

```python
# Jawaban 4
# Horizontal bar chart menggambarkan jumlah pengguna dari setiap tipe kartu

df['Card_Category'].value_counts().plot(kind='barh', title='jumlah pengguna dari setiap tipe kartu')
```

```python
# Jawaban 5
# Stacked bar chart dari pengguna dengan status perkawinan "Married", "Single", dan "Divorced" 
# dengan jenis kelamin sebagai sumbu x dan status perkawinan sebagai stacked y

nonUknown = df.loc[df['Marital_Status'] != 'Unknown'] #Uknown tidak termasuk
nonUknown.groupby(['Gender', 'Marital_Status']).size().unstack().plot(kind= 'bar', stacked=True)
```
